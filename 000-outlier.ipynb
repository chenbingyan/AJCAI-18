```python
import pandas as pd
import numpy as np
import time
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from collections import Counter
```

```python
data = pd.read_csv('./data/round1_ijcai_18_train_20180301.txt', sep=' ', encoding='utf-8')
data.head()
```

```python
data.drop_duplicates(inplace=True)
```

```python
items = data.item_id
items_counter = Counter()
for item in items:
    if item in items_counter:
        items_counter[item] += 1
    else:
        items_counter[item] = 1
```

```python
for item in items_counter:
    print(item, items_counter[item])
```
